## Distribution method: - z-score method

```python
data['SepalWidthCm'].skew()
```

```python
data.describe()
```

```python
print('Highest score: ',data['SepalWidthCm'].mean()+3*data['SepalWidthCm'].std())
print('Lowest value: ', data['SepalWidthCm'].mean()-3*data['SepalWidthCm'].std())
```

```python
#finding outliers
data[(data['SepalWidthCm']>4.35)| (data['SepalWidthCm']<1.75)]
```

```python
# trimming
new_df = data[(data['SepalWidthCm']<4.35)| (data['SepalWidthCm']>1.75)]
```

```python
new_df['SepalWidthCm']
```

```python
# cpping method

data['z_score'] = (data['SepalWidthCm']-data['SepalWidthCm'].mean())/data['SepalWidthCm'].std()
```

```python
data
```

```python
data[(data['SepalWidthCm']>4.35)|(data['SepalWidthCm']<1.75)]
```

```python
# capping
upper_limit = data['SepalWidthCm'].mean()+3*data['SepalWidthCm'].std()
lower_limit = data['SepalWidthCm'].mean()-3*data['SepalWidthCm'].std()
print('Upper_limit: ', upper_limit)
print('Lower_limit: ', lower_limit)
```

```python
data['SepalWidthCm'] = np.where(data['SepalWidthCm']>upper_limit,upper_limit,
                            np.where(data['SepalWidthCm']<lower_limit,lower_limit,
                               data['SepalWidthCm']))
```
